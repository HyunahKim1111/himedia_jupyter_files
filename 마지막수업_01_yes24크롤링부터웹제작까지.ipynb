```python
import time  # 5초간 휴식 할 때 사용
```

```python
year = list(range(2023,2007,-1))
month = list(list(range(1,13)))
page = list(range(1,10))
print(year)
print(month)
print(page)
```

```python
for yy in year:
    for mon in month if yy != yyyy else range(1, 12):
        for p in page:
            print(yy, mon, p)
```

```python
result = []
for Yr in year[:1]:
    for Mth in month if Yr != today.year else range(1,12):
        for p in page[:1]:
            print(f"{Yr}년 {Mth}월 {p}페이지 수집중", end="\r")
            url = 'https://www.yes24.com/Product/Category/MonthWeekBestSeller'
            params = dict(categoryNumber='001', pageNumber=page, pageSize=120, type='month', saleYear=year, saleMonth=month)
            try:
                r = requests.get(url, params=params)
            except Exception as e:
                print(e)
            
            soup = bs(r.text, 'lxml')
            books = soup.select_one('#yesBestList').select("li")
            
            for book in books:    
                title = book.select_one(".gd_name").text if book.select_one(".gd_name") != None else ''
                author = book.select_one(".info_row.info_pubGrp a").text # author
                pub = book.select_one(".authPub.info_pub").text if book.select_one(".authPub.info_pub") != None else ''
                pub_date = book.select_one(".authPub.info_date").text if book.select_one(".authPub.info_date") !=None else ''
                rating = book.select_one(".rating_grade > em").text if book.select_one(".rating_grade > em") != None else '' #rating
                detail_link ="https://www.yes24.com"+soup.select_one(".info_row.info_name > a").get("href")
                review_link = book.select_one(".rating_rvCount > a").get("href") if book.select_one(".rating_rvCount > a") != None else ''
                image = book.select_one(".lazy").get("src") if book.select_one(".lazy") != None else ''
                result.append({"Title":title, "Author":author, "Publisher":pub, "PubDate":pub_date, "Rating":rating, "DetailLink":detail_link, \
                                   "ReviewLink":review_link, "Image":image})
        
```

```python
# 의철씨 버전: 파람을 구지 하지 않음.
result = []
for Yr in year:
    for Mth in month if Yr != today.year else range(1,12):
        for p in page:
            print(f"{Yr}년 {Mth}월 {p}페이지 수집중", end="\r")
            url = f'https://www.yes24.com/Product/Category/MonthWeekBestSeller?categoryNumber=001&pageNumber={p}&pageSize=120&type=month&saleYear={Yr}&saleMonth={Mth}'
            try:
                r = requests.get(url)
            except Exception as e:
                print(e)
    
            soup = bs(r.text, 'lxml')
            books = soup.select('#yesBestList > li')
            
            for book in books:    
                title = book.select_one(".gd_name").text if book.select_one(".gd_name") != None else ''
                author = book.select_one(".authPub.info_auth").text.strip() if book.select_one(".authPub.info_auth") != None else''
                pub = book.select_one(".authPub.info_pub").text if book.select_one(".authPub.info_pub") != None else ''
                pub_date = book.select_one(".authPub.info_date").text if book.select_one(".authPub.info_date") !=None else ''
                rating = book.select_one(".rating_grade > em").text if book.select_one(".rating_grade > em") != None else '' #rating
                detail_link ="https://www.yes24.com"+soup.select_one(".info_row.info_name > a").get("href")
                review_link = book.select_one(".rating_rvCount > a").get("href") if book.select_one(".rating_rvCount > a") != None else ''
                image = book.select_one(".lazy").get("src") if book.select_one(".lazy") != None else ''
                result.append({"Title":title, "Author":author, "Publisher":pub, "PubDate":pub_date, "Rating":rating, "DetailLink":detail_link ,"ReviewLink":review_link, "Image":image})
            time.sleep(4)
        
```
