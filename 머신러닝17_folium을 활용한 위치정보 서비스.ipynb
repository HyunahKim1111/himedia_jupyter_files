* 각 역별 승하차정보

```python
# 2021 서울지하철 이용객 정보 정규화때문에 두개가 나눠진거다.
data = pd.read_csv("./data/seoul-metro-2021.logs.csv")
data.head()
```

```python
# 서울 지하철 역 정보 파일
station_info = pd.read_csv("./data/seoul-metro-station-info.csv")
station_info.head(10)
```

각 역 별 승하차 인원 구하기


```python
staion_sum = data.groupby("station_code").sum()
```

```python
staion_sum.head()
```

```python
staion_sum2 = staion_sum.drop('timestamp', axis=1)
```

```python
staion_sum2 # 스테이션 코드에 따른 승차인원, 하차인원
```

```python
# 지하철 역 정보파일에서 필요한 컬럼만 추출하기
```

```python
station_info.head(2)
```

```python
# 지하철 역 정보파일에서 필요한 컬럼만 추출하기
station_info = station_info[['station.code', 'line.name_sub', 'geo.latitude', 'geo.longitude']]
station_info
```

```python
# station_info이 인덱스로 바뀜.
station_info = station_info.set_index('station.code')
station_info
```

```python
staion_sum2
```

```python
# 인덱스를 이용해서 두개 합치기 : join
joined_data = staion_sum2.join(station_info)
joined_data
```

# 지도에 표시하기

```python
# 승차용 서울지도 만들기
seoul_in = folium.Map(location=[37.5, 126.98], zoom_start=12)
seoul_in
```

```python
# 히트맵 플러그인 모듈
from folium.plugins import HeatMap
```

```python
joined_data.columns
```

```python
# 히트맵 플러그인을 서울 지도에 추가하기
HeatMap(data = joined_data[['geo.latitude','geo.longitude','people_in']]).add_to(seoul_in)
seoul_in
```

```python
# 하차용 지도 만들고 하차 인원 히트맵에 추가
seoul_out = folium.Map(location=[37.55, 126.98], zoom_start=12)
# 히트맵 플러그인을 서울 지도에 추가하기
HeatMap(data = joined_data[['geo.latitude','geo.longitude','people_out']]).add_to(seoul_out)
seoul_out
```

# 시간을 추출해서 출퇴근 시간대 혼잡 역 히트맵 그리기

## timestamp를 시간 데이터로 변환하기 
## pd.to_datetime

```python
# timestamp를 시간 데이터로 변환하기 
# pd.to_datetime
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
# dt.year, month, day, hour, minute, second
data['timestamp'].dt.year
```

```python
from datetime import datetime
```

```python
today = datetime.today()
```

```python
today.year
```

```python
# 오전 9시 이전 데이터
morning_data = data[data['timestamp'].dt.hour < 9]
```

```python
morning_data
```

```python
# 퇴근용 데이터 추출
evening_data = data[data['timestamp'].dt.hour > 18]
evening_data
```

```python
morning_data.drop('timestamp', axis=1, inplace=True)
evening_data.drop('timestamp', axis=1, inplace=True)
```

```python
morning_station_sum = morning_data.groupby('station_code').sum()
evening_station_sum = evening_data.groupby('station_code').sum()
```

```python
morning_station_sum
```

```python
evening_station_sum
```

```python
morning_joined_data = morning_station_sum.join(station_info)
evening_joined_data = evening_station_sum.join(station_info)
```

```python
morning_joined_data.head()
```

```python
evening_joined_data.head()
```

```python
# 출근시간용 지도 만들고 승차 인원 히트맵 지도에 표시하기
# 출근 시간용 

seoul_morning_in = folium.Map(location=[37.55, 126.98], zoom_start=12)
# 히트맵 플러그인을 서울 지도에 추가하기
HeatMap(data = morning_joined_data[['geo.latitude','geo.longitude','people_in']]).add_to(seoul_morning_in)
seoul_morning_in
```

```python
# 출근시간용 지도 만들고 승차 인원 히트맵 지도에 표시하기
# 출근 시간용 하차

seoul_evening_out = folium.Map(location=[37.55, 126.98], zoom_start=12)
# 히트맵 플러그인을 서울 지도에 추가하기
HeatMap(data = evening_joined_data[['geo.latitude','geo.longitude','people_out']]).add_to(seoul_evening_out)
seoul_evening_out
```

```python
# 점심시간에 가장 많이 탑승/ 하차하는 곳 히트맵 11-13
lunchtime = data.loc[(data['timestamp'].dt.hour >= 11) & (data['timestamp'].dt.hour <= 13)]
```

```python
lunchtime.drop('timestamp', axis=1, inplace=True)
```

```python
lunchtime
```

```python
lunchdata_sum = lunchtime.groupby('station_code').sum()
```

```python
lunchdata_sum
```

```python
lunch_joined_data = lunchdata_sum.join(station_info)
```
