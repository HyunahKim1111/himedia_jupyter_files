```python
# tr 두개를 나누기
book_info1 = []
book_info2 = []
for i, tr in enumerate(soup.select("tr")):
    if i % 2 == 0:
        book_info1.append(tr)
    else:
        book_info2.append(tr)
```

```python
result1 = []
result2 = []
for info in book_info1:
    title = info.select_one("td.goodsTxtInfo p").text
    detail_link = 'https://www.yes24.com' + info.select_one("td.goodsTxtInfo p > a")["href"]
    author, publisher, pub_date = info.select_one("div.aupu").text.split("|")
    ori_price = info.select_one(".aupu + p").text.split("→")[0] # 원래가격
    disc_price = info.select_one(".aupu + p").text.split("→")[1].split("원")[0] # 할인가격
    num_review = info.select_one("p.review").text.split("(")[1].split("개")[0]  # 리뷰갯수
    content_review = 0
    edit_review =0
    for i in info.select_one("p.review").select("img")[:5]: # 내용리뷰
        if len(str(i)) == 66:
            content_review += 1
    for i in info.select_one("p.review").select("img")[5:]: # 편집리뷰
        if len(str(i)) == 66:
            edit_review += 1
    result1.append([title, detail_link, author, publisher, pub_date, ori_price, disc_price, num_review, content_review, edit_review])
    
for info2 in book_info2:
    preview = info2.select_one("p.read").text
    result2.append(preview)
    
```

```python
#result2개를 합치기
# item의 길이가 10이니까 11번에 insert를 해주면 돼.
for index, item in enumerate(result1):
    print(index, len(item))
    result1[index].insert(11, result2[index])
result1
```

# zip 함수 사용하기: 여러개를 같이 묶어서 for문 돌릴 때 사용

```python
final_result = []
for info, preview in zip(book_info1, book_info2):
    title = info.select_one("td.goodsTxtInfo p").text
    detail_link = 'https://www.yes24.com' + info.select_one("td.goodsTxtInfo p > a")["href"]
    author, publisher, pub_date = info.select_one("div.aupu").text.split("|")
    ori_price = info.select_one(".aupu + p").text.split("→")[0] # 원래가격
    disc_price = info.select_one(".aupu + p").text.split("→")[1].split("원")[0] # 할인가격
    num_review = info.select_one("p.review").text.split("(")[1].split("개")[0]  # 리뷰갯수
    content_review = 0
    edit_review =0
    for i in info.select_one("p.review").select("img")[:5]: # 내용리뷰
        if len(str(i)) == 66:
            content_review += 1
    for i in info.select_one("p.review").select("img")[5:]: # 편집리뷰
        if len(str(i)) == 66:
            edit_review += 1
    preview = preview.select_one("p.read").text
    final_result.append([title, detail_link, author, publisher, pub_date, ori_price, disc_price, num_review, content_review, edit_review, preview])
```
