```python
import requests
year = list(range(2023,2007,-1))
month = list(list(range(1,13)))
page = list(range(1,10))
review_links = []
for Yr in year:
    for Mth in month if Yr != today.year else range(1,12):
        for p in page:
            url = f'https://www.yes24.com/Product/Category/MonthWeekBestSeller?categoryNumber=001&pageNumber={p}&pageSize=120&type=month&saleYear={Yr}&saleMonth={Mth}'
            r = requests.get(url)
            soup = bs(r.text, 'lxml')
            books = soup.select('#yesBestList > li')
            for book in books:
                review_link = book.select_one(".rating_rvCount > a").get("href") if book.select_one(".rating_rvCount > a") != None else ''
                review_links.append({"ReviewLink":review_link})
                print(f"{Yr}년 {Mth}월 {p}페이지 수집중", end="\r")
            time.sleep(3)
```

```python
# 스크롤 내려서 눈에 보이는 숫자까지 수집완료! 다음은 > 버튼을 누르는 건데..
from selenium import webdriver
test3 = []
for idx, link in enumerate(review_links_df['ReviewLink'][:2]):
    print(f"{idx}번째 link 수집중") 
    driver.get(link)
    time.sleep(3)
    soup = bs(driver.page_source, 'lxml')
    title = soup.select_one('.gd_name').text
    # 수직으로 500px스크롤을 내려라
    driver.execute_script("window.scrollBy(0, 500);") 
    driver.find_element(By.CSS_SELECTOR,"#purchase").click()
    
    page=1
    while True:
        try:
            soup = bs(driver.page_source, 'lxml')
            for idx, review in enumerate(soup.select_one('#infoset_reviewContentList').select('.reviewInfoBot.origin > .review_cont')):
                test_review = review.text.replace('\n\xa0\n','').strip()
                print(f"{idx}번째 리뷰수집중") 
                test3.append({"Title":title, "테스트리뷰":test_review})
                time.sleep(3)
            driver.find_element(By.CSS_SELECTOR, 'div.yesUI_pagenS > strong.num + a.num').click()
            print(f"{page}페이지 넘김")
            time.sleep(3)
            page+=1
        except Exception as e:
            print("버튼없음",e)
            break
            
```

```python
# 다음은 > 버튼을 누르는 건데..
from selenium import webdriver
test3 = []
for idx, link in enumerate(review_links_df['ReviewLink'][3:5]):
    print(f"{idx}번째 link 수집중") 
    driver.get(link)
    time.sleep(3)
    soup = bs(driver.page_source, 'lxml')
    title = soup.select_one('.gd_name').text
    # 수직으로 500px스크롤을 내려라
    driver.execute_script("window.scrollBy(0, 500);") 
    time.sleep(5)
    driver.find_element(By.CSS_SELECTOR,"#purchase").click()
    
    page=1
    while True:
        try:
            if page ==1:
                print(f"{page}페이지")
                soup = bs(driver.page_source, 'lxml')
                for idx, review in enumerate(soup.select_one('#infoset_reviewContentList').select('.reviewInfoBot.origin > .review_cont')):
                    test_review = review.text.replace('\n\xa0\n','').strip()
                    print(f"{idx}번째 리뷰수집중") 
                    test3.append({"Title":title, "Review":test_review})
                    time.sleep(3)
                driver.find_element(By.CSS_SELECTOR, '.review_sortLft > div.yesUI_pagenS strong + a.num').click()
                time.sleep(3)
                page+=1
                
            elif page % 10 != 0:
                print(f"{page}페이지")
                soup = bs(driver.page_source, 'lxml')
#                 driver.execute_script("window.scrollBy(0, 500);") 
#                 time.sleep(3)
                for idx, review in enumerate(soup.select_one('#infoset_reviewContentList').select('.reviewInfoBot.origin > .review_cont')):
                    test_review = review.text.replace('\n\xa0\n','').strip()
                    print(f"{idx}번째 리뷰수집중") 
                    test3.append({"Title":title, "Review":test_review})
                    time.sleep(3)
                driver.find_element(By.CSS_SELECTOR, '.review_sortLft > div.yesUI_pagenS strong + a.num').click()
                time.sleep(3)
                page+=1
                
            elif page % 10 == 0:
                print(f"{page}페이지")
                soup = bs(driver.page_source, 'lxml')
#                 driver.execute_script("window.scrollBy(0, 500);") 
#                 time.sleep(3)
                for idx, review in enumerate(soup.select_one('#infoset_reviewContentList').select('.reviewInfoBot.origin > .review_cont')):
                    test_review = review.text.replace('\n\xa0\n','').strip()
                    print(f"{idx}번째 리뷰수집중") 
                    test3.append({"Title":title, "Review":test_review})
                    time.sleep(3)
                driver.find_element(By.CSS_SELECTOR, 'div.yesUI_pagenS > strong.num + a.bgYUI.next').click()
                time.sleep(3)
                page+=1
                
        except Exception as e:
            print("버튼없음",e)
            break
            
```
