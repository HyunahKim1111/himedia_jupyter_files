```python
hotel_info = []
for i in data:
    for j in i['data']['body']['searchResults']['results'][:2]:
        hotel_id = j['id']
        name = j['name']
        starRating = j['starRating']
        rating = j['guestReviews']['rating']
        total = j['guestReviews']['total']
        current = j['ratePlan']['price']['current']
        if 'old' in j['ratePlan']['price']:
            old = j['ratePlan']['price']['old']
        else:
            old = j['ratePlan']['price']['current']
        locality = j['address']['locality']
        streetAddress = j['address']['streetAddress']
        lat = j['coordinate']['lat']
        lon = j['coordinate']['lon']
    
        hotel_info.append([hotel_id, name, starRating, rating, total, current, old, locality, streetAddress, lat, lon])
        print([hotel_id],[name],[starRating], [rating], [total], [current], [old], [locality], [streetAddress], [lat], [lon])
        







# lat = data[0]['data']['body']['searchResults']['results'][0]['coordinate']['lat']
# lon = data[0]['data']['body']['searchResults']['results'][0]['coordinate']['lon']
```

```python
hotel_info = []
for i in data:
    for j in i['data']['body']['searchResults']['results'][:2]:
        hotel_id = j.get('id', 0)
        name = j.get('name', "unknown")
        starRating = j.get('starRating', 0)
        
        #오류잡는 코드 : 예외처리 1
        if 'guestReviews' in j:
            rating = j['guestReviews'].get('rating', 0)
        else:
            rating = 0
            
        #오류잡는 코드 : 예외처리 2
        current = j['ratePlan']['price'].get('current', 0) if 'ratePlan' in j else ()
        
        #오류잡는 코드 : 예외처리 3
        try: # 정상이면 그냥 실행
            old = j['ratePlan']['price'].get('old', 0) if 'ratePlan' in j else ()
        except Exception as e: # 모든 오류를 잡아서 출력해줘. 오류가 있으면 오류 난 내용을 e에 저장하고 e를 프린트해라
            print(e)
        finally: # 오류 난 부분에 0을 붙여서 실행해라. else와 같은 작용.
            old = 0
            
        locality = j['address'].get('locality', 'unknown')
        streetAddress = j['address'].get('streetAddress', 'unknown')
        lat = j['coordinate'].get('lat', 0)
        lon = j['coordinate'].get('lon',0)
    
        hotel_info.append([hotel_id, name, starRating, rating, total, current, old, locality, streetAddress, lat, lon])
        print([hotel_id],[name],[starRating], [rating], [total], [current], [old], [locality], [streetAddress], [lat], [lon])
```
