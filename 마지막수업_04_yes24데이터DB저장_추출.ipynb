```python
def fromyes24db(dbid, dbpw, db_name, table_name):
    from sqlalchemy import create_engine
    import pymysql
    import yes24dbinfo
    pymysql.install_as_MySQLdb() # pymysql을 사용하려면 필수로 같이 써줘야 해. 특히 sqlalchemy와 연동해서 쓰러면
    engine = create_engine("mysql+mysqldb://{dbid}:{dbpw}@192.168.0.53:3306/{db_name}") #워크벤트 켜고 접속하기 위해서 작성
    #engine = create_engine("mysql+mysqldb://접속할 때 사용할 아이디:password")
    conn = engine.connect() #접속 버튼 누름. 쿼리 창이 열림
    yes24 = pd.read_sql("SELECT * FROM {table_name}", con=conn)
    yes24 = pd.read_sql(f"SELECT * FROM {table_name} WHERE ""={}, con=conn)
    return yes24
```
