```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# company guide사이트
url = "https://comp.fnguide.com/SVO2/ASP/SVD_main.asp?pGB=1&gicode=A005930&cID=&MenuYn=Y&ReportGB=&NewMenuID=11&stkGb=&strResearchYN="
# urlopen함수로 해당 웹 페이지를 열고 변수를 html에 저장
html = urlopen(url)
# BeautifulSoup 함수를 이용해 html의 내용을 파싱
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
# 날짜
date1 = soup.find("span", {"class":"date"})
print(date1.text)
date2 = date1.text
date = date2.replace('[','').replace(']','').replace('/','-')
print(date)
```

```python
#종목이름
corp_name1 = soup.find_all("h1", {"id":"giName"})
print(corp_name1)
corp_name = corp_name1[0].text
print(corp_name)
```

```python
#종목 코드
code1 = soup.find_all("div", {"class":"corp_group1"})
print(code1)
code2 = code1[0].find("h2")
print(code2)
# text 메서드를 사용해 태그를 제외하고 텍스트 데이터만 출력
code = code2.text
print(code)
```

```python
#주가
stock_price1 = soup.find("span", {"id":"svdMainChartTxt11"})
print(stock_price1)
stock_price2 = stock_price1.text
print(stock_price2)
stock_price = int(stock_price2.replace(',','').strip())
print(stock_price)
```

```python
# 외국인 보유 비중
fgn_own_ratio1 = soup.find("span", {"id":"svdMainChartTxt12"})
print(fgn_own_ratio1)
fgn_own_ratio = float(fgn_own_ratio1.text)
print(fgn_own_ratio)
```

```python
#상대 수익률
rel_return1 = soup.find("span", {"id":"svdMainChartTxt13"})
print(rel_return1)
rel_return = float(rel_return1.text)
print(rel_return)
```

```python
# 'PER', '12MPER', '업종 PER', 'PBR', '배당수익률' 테이블
#<dl> 태그는 Definition List의 약자로 용어의 목록을 만드는 태그
#<dt>태그는 Definition Term의 약자로 용어나 이름을 나타낼 때 사용
#<dd>태그는 Definition Description의 약자로 해당 용어를 설명할 때 사용.
up_list = soup.find("div",{"class":"corp_group2"})
print(up_list)
```

```python
dd = up_list.find_all("dd")
print(dd)
```

```python
#PER(Rrice Earning Ratio): 주가수익비율
per = float(dd[1].text)
print(per)
```

```python
# 12M PER: 12개월 뒤의 예상 주가수익비율
per_12m = float(dd[3].text)
print(per_12m)
```

```python
# 업종 PER
per_ind = float(dd[5].text)
print(per_ind)
```

```python
# PBR(Price to Book Ratio):주가순자산비율, 주가를 주당순자산으로 나눈 것
pbr = float(dd[7].text)
print(pbr)
```

```python
# 배당수익률
div_yid1 = dd[9].text
print(div_yid1)
div_yid2 = div_yid1.replace('%','')
print(div_yid2)
div_yid = float(div_yid2)
print(div_yid)
```

```python
#시세현황 테이블 웹 크롤링
table1 = soup.find("div",{"id":"div1"})
print(table1)
table2 = table1.find_all("td")
print('-----------------------------------------')
print(table2)
```

```python
#거래량
volume1 = table2[1].text
print(volume1)
volume = int(volume1.replace(',','').strip())
print(volume)
```

```python
# 거래대금
trans_price1 = table2[3].text
print(trans_price1)
trans_price = int(trans_price1.replace(',','').strip())
print(trans_price)
```

```python
# 시가총액(우선주 포함)
mk_cpt_pfr1 = table2[6].text
print(mk_cpt_pfr1)
mk_cpt_pfr = int(mk_cpt_pfr1.replace(',','').strip())
print(mk_cpt_pfr)
```

```python
# 시가총액(보통주)
mk_cpt_cm1 = table2[8].text
print(mk_cpt_cm1)
mk_cpt_cm = int(mk_cpt_cm1.replace(',','').strip())
print(mk_cpt_cm)
```

```python
# 결과 모음
res = [date, corp_name, code, stock_price, fgn_own_ratio, rel_return, per, per_12m, per_ind, pbr, div_yid, volume, trans_price, mk_cpt_pfr, mk_cpt_cm]
print(res)
```
