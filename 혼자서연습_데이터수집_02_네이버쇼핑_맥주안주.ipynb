```python
# selenium 4
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.set_window_size(1920, 1080) # 웹 브라우저 해상도 조절
keyword = input("검색할 제품 입력: ")
driver.get(f"https://search.shopping.naver.com/search/all?frm=NVSHATC&origQuery={keyword}&pagingSize=20&productSet=total&query=%EB%A7%A5%EC%A3%BC%EC%95%88%EC%A3%BC&sort=rel&timestamp=&viewType=list")
```

```python
snacks_for_beer = []
for page in range(2,11):
    print(f"{page}페이지")
    y = 0 # 마우스 스크롤 시작 지점
    y_step = 1000 # 마우스 스크롤 길이
    for i in range(1, 6):
        y =  y + y_step
        script = f"window.scrollTo({0},{y})"  # 자바 스크립트로 위치 지정
        driver.execute_script(script)         # 스크롤 실행
        time.sleep(2)
    page_html = driver.page_source 
    soup = BeautifulSoup(page_html, "lxml")
    snacks_for_beer.append(soup)
    driver.find_element(By.CSS_SELECTOR, f"#content > div.style_content__xWg5l > div.pagination_pagination__fsf34 > div > a:nth-child({page})").click() # 다음페이지 넘김 버튼 클릭
    time.sleep(2)
```

```python
# 의철씨 방법
pages = 2
snacks_for_beer = []
while pages <= 10:
    page_html = driver.page_source
    soup = BeautifulSoup(page_html, "lxml")
#     print(f"현재 버튼숫자{nth_child}")
    print(f"현재 페이지{pages}")
#     data-nclick="N=a:pag.page,i:{pages}
    driver.find_element(By.CSS_SELECTOR,f"#content > div.style_content__xWg5l > div.pagination_pagination__fsf34 > div > a:nth-child({pages})").click()
    pages += 1
    time.sleep(3)
    snacks_for_beer.append(soup)
    
#     if pages % 10 == 0:
#         nth_child = 4
#         driver.find_element(By.CSS_SELECTOR,".css-15xqn45-defaultStyle-icon_057-Icon").click()
```

```python
pages = 2
snacks_for_beer = []

while pages <= 10:
    page_html = driver.page_source
    soup = BeautifulSoup(page_html, "lxml")

    print(f"현재 페이지{pages}")

    # data-nclick 속성을 사용하여 페이지 숫자를 가져오기
    a_tag = soup.find('a', class_='pagination_btn_page___ry_S')
    if a_tag:
        data_nclick = a_tag['data-nclick']
        number = data_nclick.split(':')[-1].strip('} ')

        print(f"현재 페이지 번호: {number}")

    # 페이지 전환을 위해 버튼 클릭
    driver.find_element(By.CSS_SELECTOR, f"#content > div.style_content__xWg5l > div.pagination_pagination__fsf34 > div > a:nth-child({pages})").click()

    pages += 1
    time.sleep(3)
    snacks_for_beer.append(soup)
```
