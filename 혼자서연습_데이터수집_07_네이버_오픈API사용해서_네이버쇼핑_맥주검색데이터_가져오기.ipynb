```python
# 네이버 검색 오픈 API 사용
import os
import sys
import urllib.request
page = 1 #초기화
start_num = 1
n_item = 100
beers_in_Naver = []
while True:
    client_id = "y5NUrbugaum6JAS_oEYk"
    client_secret = "oVPm8scYB5"
    encText = urllib.parse.quote("맥주") # 서버에 요청을 할 때 uft-8로 변환
    url = "https://openapi.naver.com/v1/search/shop.json?query=" + encText
    url += f"&display={n_item}&start={start_num}&sort=sim"
    print(url) 

    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)

    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        print(response_body.decode('utf-8'))
    else:
        print("Error Code:" + rescode)
    data = response_body.decode('utf-8') #결과를 data라는 변수에 담음
    data = json.loads(data)
    total_page = data['total'] // n_item + 1 #923 // 100 = 9.23 그래서 +1이 필요. total_page:10
    beers_in_Naver.append(data) #json모드 import하고 json데이터 파일로 변환.
    
    if page == 10:
        break
    
    page += 1
    start_num += 100
```
