```python
# HTML 태그 제거하고 한글만 추출하는 함수
def extract_korean(text):
    soup = BeautifulSoup(text, 'html.parser')
    text_without_tags = soup.get_text()
    korean_text = ''.join(char for char in text_without_tags if char.isalpha())  # 한글만 추출하려면 .isalpha() 사용
    return korean_text

# 데이터프레임의 'text' 열에 적용
#df['korean_text'] = df['text'].apply(extract_korean)

# 결과 출력
#print(df['korean_text'])
```
