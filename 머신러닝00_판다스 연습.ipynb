# 판다스 데이터 프레임에서 조건에 맞는 행만 가져오기(중요!)
* 데이터프레임[데이터프레임['컬럼명'] 조건식]
=> 조건식이 참인 것만 가져옴
* >, < >=, <=, ==, !=, &(and), |(or)

```python
# 조건이 1개인 경우
data[data['범죄율']<0.05]
```

```python
# 조건이 2개인 경우에는 각 조건을 ()로 묶어줘야 한다.
data[(data['범죄율']>0.05) & (data['범죄율']<0.1)]
```

```python
# 조건이 3개인 경우 : 범죄율 0.05 미만, 가구당 평균방수가 4개 이상, 하위계층비율 3이하인 집                                        
data[(data['범죄율']<0.05) & (data['가구당평균방수']>=4) & (data['하위계층비율']<=3)]
```

```python
# 앞에 있는 조건을 만족하면서 원하는 컬럼 확인하기.
# loc : 조건식을 넣어서 내가 원하는 컬럼을 조회할 때는 .loc로 한다. (R에서 where와 비슷) 단, 전체컬럼을 조회할 때는 .loc를 안 써도 돼.
data.loc[(data['범죄율']<0.05) & (data['가구당평균방수']>=4) & (data['하위계층비율']<=3),['가구당평균방수','주택가격'] ]
```

```python
data.loc[(data['범죄율']<0.05) & (data['가구당평균방수']>=4) & (data['하위계층비율']<=3) ]
```
