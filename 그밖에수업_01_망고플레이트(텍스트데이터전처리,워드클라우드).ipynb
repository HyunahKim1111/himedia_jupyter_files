```python
# 'ID' 컬럼을 기준으로 그룹화하여 그룹별로 데이터프레임 생성
grouped_dataframes = dict(tuple(data.groupby('ID')))

# 각각의 이름별 데이터프레임 출력
for ids, data in grouped_dataframes.items():
    print(f"이름: {ids}")
    print(data)
#     print(len(f"이름: {ids}"))
```

```python
# 각각의 ID별 데이터프레임을 함수화하여 globals()로 선언
def create_id_dataframes():
    for ids, data in grouped_dataframes.items():
        globals()[f"ID_{ids}_DataFrame"] = data
        
# 함수 실행
create_id_dataframes()

# # 생성된 함수를 이용해 각 ID별 데이터프레임 확인
# print(ID_1_DataFrame)
# print(ID_2_DataFrame)
# print(ID_3_DataFrame)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# 'ID' 컬럼을 기준으로 그룹화하여 그룹별로 데이터프레임 생성 (예시)
grouped_dataframes = dict(tuple(data.groupby('ID')))

# 각 ID별로 텍스트 데이터를 합치기 위한 빈 문자열 생성
text_combined = ''

# 각 데이터프레임에서 'text' 컬럼을 가져와서 텍스트를 합침
for ids, data in grouped_dataframes.items():
    text_combined += ' '.join(data['리뷰내용']) + ' '

# WordCloud 생성
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_combined)
# 폰트위치
font = '/usr/share/fonts/opentype/noto/NotoSansCJK-Black.ttc'

# 워드클라우드 출력
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
